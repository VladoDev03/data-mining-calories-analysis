# Calories Analysis

## Data Cleaning

```python
def normalize_column_names(df):
    df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
    return df
```

```python
def binary_encode(df, column, val1 = 'yes', val2 = 'no'):
    df[column] = df[column].str.lower().map({val1.lower(): 1, val2.lower(): 0})
```

```python
calories = pd.read_csv("data/calories.csv")
```

```python
calories
```

```python
calories = normalize_column_names(calories)
```

```python
calories.gender.value_counts()
```

```python
binary_encode(calories, 'gender', 'male', 'female')
```
